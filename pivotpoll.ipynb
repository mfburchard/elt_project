```python
import pandas as pd
```

```python
polls = pd.read_csv('Resources/poll_numbers.csv')
polls = polls.rename(columns = {'name' : 'candidate'})

polls
```

```python
polls = polls.pivot(index = 'candidate', columns='date', values='percent')
polls
```

```python
polls['candidate_id'] = [1,2,3,4,5,6,7]
polls
```

```python
polls = polls.reset_index()
polls = polls.set_index(['candidate_id', 'candidate'])
polls
```

```python
polls.to_csv('Final/poll_count.csv')
```
